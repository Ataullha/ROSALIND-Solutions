# ba1a - Compute the Number of Times a Pattern Appears in a Text

```python
def ba1a(s,t):
    cnt = 0
    for i in range(len(s)-len(t)+1):
        if s[i:i+len(t)] == t:
            cnt += 1
    return cnt

ba1a('GCGCG','GCG')
```

# ba1b - Find the Most Frequent Words in a String

```python
def ba1b(s,k):
    m = {}
    for i in range(len(s)-k+1):
        if s[i:i+k] not in m:
            m[s[i:i+k]] = 0
        m[s[i:i+k]] += 1
    for i in m:
        if m[i]==max(m.values()):
            print(i,end=' ')

ba1b('ACGTTGCATGTCGCATGATGCATGAGAGCT',4)
```

# ba1c - Find the Reverse Complement of a String 

```python
def ba1c(s):
    t = ''
    for i in s:
        if i=='A':
            t = t + 'T'
        elif i=='T':
            t = t + 'A'
        elif i=='G':
            t = t + 'C'
        elif i=='C':
            t = t + 'G'
    return t[::-1] 
    
ba1c('AAAACCCGGT')
```

# ba1d - Find All Occurrences of a Pattern in a String

```python
def ba1d(t,s):
    for i in range(len(s)-len(t)+1):
        if s[i:i+len(t)]==t:
            print(i,end=' ')

ba1d('ATAT','GATATATGCATATACTT')
```

# *ba1e - Find Patterns Forming Clumps in a String

```python
def ba1e(s,k,L,t):
    kmers = []
    for i in range(len(s)-k+1):
        kmers.append(s[i:i+k])
    # print(kmers)
    allpss = []
    for i in range(len(s)-L+1):
        allpss.append(s[i:i+L])
    # print(allpss)

    ans = set()

    for i in allpss:
        for j in kmers:
            if i.count(j)>=t:
                ans.add(j)
    for i in ans:
        print(i,end=' ')

ba1e('CGGACTCGACAGATGTGAAGAAATGTGAAGACTGAGTGAAGAGAAGAGGAAACACGACACGACATTGCGACATAATGTACGAATGTAATGTGCCTATGGC',5,75,4)
```

# *ba1f - Find a Position in a Genome Minimizing the Skew

```python
import matplotlib.pyplot as plt

def ba1f(s):
    m = {}
    g = 0
    c = 0
    for i in range(len(s)):
        if s[i]=='G':
            g += 1
        elif s[i]=='C':
            c += 1
        temp = g-c
        m[i]=temp
    # display(m)
    for i in m:
        if m[i]==min(m.values()):
            print(i+1,end=' ')
    
    plt.figure(figsize=(12,4))
    plt.plot(list(m.keys()),list(m.values()))
    plt.xticks(list(m.keys()),s)
    plt.show()


ba1f('CCTATCGGTGGATTAGCATGTCCCTGTACGTTTCGCCGCGAACTAGTTCACACGGCTTGATGGCAAATGGTTTTTCCGGCGACCGTAATCGTCCACCGAG')
```

# ba1g - Compute the Hamming Distance Between Two Strings

```python
def ba1g(s,t):
    cnt = 0
    for i,j in zip(s,t):
        cnt += (i!=j)
    return cnt

ba1g('GGGCCGTTGGT','GGACCGTTGAC')
```

# ba1h - Find All Approximate Occurrences of a Pattern in a String

```python
def hd(s,t):
    cnt = 0
    for i,j in zip(s,t):
        cnt += (i!=j)
    return cnt

def ba1h(t,s,d):
    for i in range(len(s)-len(t)+1):
        if hd(s[i:i+len(t)],t)<=d:
            print(i,end=' ')

ba1h('ATTCTGGA','CGCCCGAATCCAGAACGCATTCCCATATTTCGGGACCACTGGCCTCCACGGTACGGACGTCAATCAAATGCCTAGCGGCTTGTGGTTTCTCCTACGCTCC',3)
```

# **ba1i - Find the Most Frequent Words with Mismatches in a String

```python
from itertools import product

def hd(s,t):
    cnt = 0
    for i,j in zip(s,t):
        cnt += (i!=j)
    return cnt

def mm(s,k,d):
    l = [''.join(i) for i in list(product('ACGT',repeat=k))]
    ans = []
    for i in l:
        if hd(i,s)<=d:
            ans.append(i)
    return ans
def b1ai(s,k,d):
    m = {}
    for i in range(len(s)-k+1):
        l = mm(s[i:i+k],k,d)
        # print(l)
        for i in l:
            if i not in m:
                m[i]=0
            m[i] += 1
    # display(m)
    for i in m:
        if m[i]==max(m.values()):
            print(i,end=' ')


b1ai('ACGTTGCATGTCGCATGATGCATGAGAGCT',4,1)
```

# BA1J - Find Frequent Words with Mismatches and Reverse Complements

```python
from itertools import product

def rc(s):
    t = ''
    for i in s:
        if i=='A':
            t = t + 'T'
        elif i=='T':
            t = t + 'A'
        elif i=='G':
            t = t + 'C'
        elif i=='C':
            t = t + 'G'
    return t[::-1]

def hd(s,t):
    cnt = 0
    for i,j in zip(s,t):
        cnt += (i!=j)
    return cnt

def mm(s,k,d):
    l = [''.join(i) for i in list(product('ACGT',repeat=k))]
    ans = []
    for i in l:
        if hd(i,s)<=d:
            ans.append(i)
    return ans
def b1ai(s,k,d):
    m = {}
    for i in range(len(s)-k+1):
        l = mm(s[i:i+k],k,d)
        # print(l)
        for i in l:
            if i not in m:
                m[i]=0
            if rc(i) not in m:
                m[rc(i)] = 0
            m[i] += 1
            m[rc(i)] += 1
    # display(m)
    for i in m:
        if m[i]==max(m.values()):
            print(i,end=' ')


b1ai('ACGTTGCATGTCGCATGATGCATGAGAGCT',4,1)
```

# ba1k - Generate the Frequency Array of a String

```python
from itertools import product

def ba1k(s,k):
    l = [''.join(i) for i in list(product('ACGT',repeat=k))]
    m = {}
    for i in l:
        m[i]=0
    for i in range(len(s)-k+1):
        m[s[i:i+k]] += 1
    # display(m)
    for i in m:
        print(m[i],end=' ')
    
ba1k('ACGCGGCTCTGAAA',2)
```

# ba1l - Implement PatternToNumber

```python
def ba1l(s):
    m = {'A':'00', 'C':'01', 'G':'10', 'T':'11'}
    t = ''
    for i in s:
        t = t + m[i]
    return int(t,2)

ba1l('AGT')
```

# ba1m - Implement NumberToPattern

```python
def ba1m(n,k):
    l = []
    m = {0:'A',1:'C',2:'G',3:'T'}
    while(n):
        l.append(n%4)
        n //= 4
    while len(l)<k:
        l.append(0)
    for i in l[::-1]:
        print(m[i],end='')
ba1m(45,4)
```
