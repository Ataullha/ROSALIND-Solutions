# ba4b - Find Substrings of a Genome Encoding a Given Amino Acid String

```python
def ba4b(dna,pattern):
    k = len(pattern)*3
    ans = []
    for i in range(len(dna)+k-1):
        kmer = dna[i:i+k]
        pep = str(Seq.Seq(kmer).transcribe().translate())
        pepr = str(Seq.Seq(kmer).reverse_complement().transcribe().translate())
        if pep==pattern or pepr==pattern:
            ans.append(kmer)
    return ans

dna = 'ATGGCCATGGCCCCCAGAACTGAGATCAATAGTACCCGTATTAACGGGTGA'
pattern = 'MA'
for i in ba4b(dna,pattern):
    print(i)
```

# ba4c - Generate the Theoretical Spectrum of a Cyclic Peptide

```python
def ba4c(pept):
    ans = []
    res = []
    m = {
        'G':57,'A':71,'S':87,'P':97,'V':99,'T':101,'C':103,'I':113,'L':113,'N':114,
        'D':115,'K':128,'Q':128,'E':129,'M':131,'H':137,'F':147,'R':156,'Y':163,'W':186
    }
    for k in range(1,len(pept)//2):
        for i in range(len(pept)//2):
            s = pept[i:i+k]
            ans.append(s)
    for i in ans:
        cnt=0
        # print(i)
        for j in i:
            cnt+=m[j]
            # print(j,end=' ')
        res.append(cnt)
    return sorted(res)
        

pept = 'LEQN'
print(0,end=' ')
for i in ba4c(pept+pept):
    print(i,end=' ')
```
